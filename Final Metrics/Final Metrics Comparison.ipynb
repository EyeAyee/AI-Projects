```python
# Plotting line charts with dotted lines
metrics = combined_df['Metric'].unique()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))

for ax, metric in zip(axes.flatten(), metrics):
    subset = combined_df[combined_df['Metric'] == metric]
    subset.plot(kind='line', x='Model Type', y='Single Model', marker='o', linestyle=':', ax=ax, title=metric, label='Single Model')
    subset.plot(kind='line', x='Model Type', y='Combined Model', marker='o', linestyle=':', ax=ax, label='Combined Model')
    ax.set_ylabel(metric)

plt.tight_layout()
plt.savefig('combined_metrics_line_charts_ANN.png')
plt.show()
```

```python
# Save individual charts
for metric in metrics:
    plt.figure()
    subset = combined_df[combined_df['Metric'] == metric]
    plt.plot(subset['Model Type'], subset['Single Model'], marker='o', linestyle=':', label='Single Model')
    plt.plot(subset['Model Type'], subset['Combined Model'], marker='o', linestyle=':', label='Combined Model')
    plt.title(metric)
    plt.ylabel(metric)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'{metric}_line_chart_ANN.png')

plt.show()
```
