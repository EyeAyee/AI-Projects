```python
import pandas as pd
import numpy as np

# Load your dataset (replace with your own data)
# For demonstration, let's assume 'df' is your DataFrame

# Calculate the correlation matrix
corr_matrix = corr_data.corr().abs()

# Select the upper triangle of the correlation matrix
upper_triangle = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Find features with correlation greater than 0.90 or less than -0.90
highly_correlated_features = [col for col in upper_triangle.columns if any(
    upper_triangle[col] > 0.90)]
```
